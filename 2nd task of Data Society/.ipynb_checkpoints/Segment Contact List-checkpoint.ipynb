```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_excel("Fastbase_business_contacts_excel_699_HVVPQG.xlsx")
```

```python
df
```

```python
df1 = df.drop(['Contact Name','Address','City','State','Zip Code','Country','Fax Number','Web Site','Revenue'],axis='columns')
df1.shape
```

```python
df1.describe()
```

```python
if df1.isnull().sum().any()>0:
    print("True")
```

```python
df1.isnull()
```

```python
df1['Industry'] = df1['Industry'].replace(['Pharma'],'Healthcare')
```

```python
df1['Industry'] = df1['Industry'].replace(['Research'],'Healthcare')
```

```python
df1['Industry'] = df1['Industry'].replace(['Consulting'],'Professional Services')
```

```python
df1
```

```python
df1['Industry'].unique()
```

```python
print(df1['Employees'].unique())
```

```python
df2=df1
```

```python
df2
```

```python
df2['Employees_New']=df2['Employees']
```

```python
df2
```

```python
df2.Employees_New = df2.Employees_New.fillna('10000000000')
```

```python
df2.isnull().sum()
```

```python
df2
```

```python
def convert_employees(x):
    if '-' in x:
        tokens = x.split('-')
        return (float(tokens[1]))
    elif 'k' in x:
        tokens =x.split('k')
        return float(tokens[0])*1000
    else:
        return float(x)
```

```python
df2.Employees_New = df2.Employees_New.apply(convert_employees)
```

```python
df2
```

```python
df2.drop(df2[df2['Employees_New'] > 500].index, inplace = True)
```

```python
df2
```

```python
file_name = 'less than five hundred.xlsx'
df2.to_excel(file_name)
```
